## Correlaciones

### -------------------------------------------------------------------------------

### Para los dos días

```python
# Gráfico de Dispersión
fig, ax = plt.subplots(1, 1, figsize=(10,8))
# Azul
ax.scatter(x= data['time'], y= data['DO_YSI_80ft(mg/L)'], alpha= 0.8)
# Naranja
ax.scatter(x= data['time'], y= data['DO_YSI_1100ft(mg/L)'], alpha= 0.8)
ax.set_xlabel('Time')
ax.set_ylabel('DO (mg/L)')
```

```python
# Cálculo de correlaciones
print("-------------------------------------------")
print("     Para los dos días     ")
print("-------------------------------------------")
# Tiempo VS DO 80ft (mg/L)
print("Tiempo VS DO 80ft (mg/L)")
print('Correlación Pearson: ', data['time'].corr(data['DO_YSI_80ft(mg/L)'], method='pearson'))
print('Correlación Spearman: ', data['time'].corr(data['DO_YSI_80ft(mg/L)'], method='spearman'))
print('Correlación Kendall: ', data['time'].corr(data['DO_YSI_80ft(mg/L)'], method='kendall'))
print("-------------------------------------------")

# Tiempo VS DO 80ft (mg/L)
print("Tiempo VS DO 1100ft (mg/L)")
print('Correlación Pearson: ', data['time'].corr(data['DO_YSI_1100ft(mg/L)'], method='pearson'))
print('Correlación Spearman: ', data['time'].corr(data['DO_YSI_1100ft(mg/L)'], method='spearman'))
print('Correlación Kendall: ', data['time'].corr(data['DO_YSI_1100ft(mg/L)'], method='kendall'))
print("-------------------------------------------")

# Tiempo VS Difference 
print("Tiempo VS Difference")
print('Correlación Pearson: ', data['time'].corr(data['difference_DO_YSI'], method='pearson'))
print('Correlación Spearman: ', data['time'].corr(data['difference_DO_YSI'], method='spearman'))
print('Correlación Kendall: ', data['time'].corr(data['difference_DO_YSI'], method='kendall'))
```
