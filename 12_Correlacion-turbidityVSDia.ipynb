```python
# Se importan las librerias
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

### Se cargan los datos

```python
data = pd.read_csv("data/data_all.csv", sep=";")

data
```

### Preprocesamiento

```python
# Se buscan registros NaN para eliminarlos
print('Columna      Cantidad NaN')
print(data.isnull().sum(axis = 0))
print(data.shape)
```

```python
# Se eliminan los registros NaN del DataFrame
data = data.dropna()
```

```python
# Se genera las estadísticas del DataFrame
data.describe()
```

## Correlaciones

### -------------------------------------------------------------------------------

### Para los dos días

```python
# Gráfico de Dispersión
fig, ax = plt.subplots(1, 1, figsize=(10,8))
# Azul
ax.scatter(x= data['time'], y= data['turbidity_YSI_80ft(NTU+)'], alpha= 0.8)
# Naranja
ax.scatter(x= data['time'], y= data['turbidity_YSI_1100ft(NTU+)'], alpha= 0.8)
ax.set_xlabel('Time')
ax.set_ylabel('Turbidity (NTU)')
```

```python
# Cálculo de correlaciones
print("-------------------------------------------")
print("     Para las dos días     ")
print("-------------------------------------------")
# Tiempo VS Turb 80ft (NTU)
print("Tiempo VS Turb 80ft (NTU)")
print('Correlación Pearson: ', data['time'].corr(data['turbidity_YSI_80ft(NTU+)'], method='pearson'))
print('Correlación Spearman: ', data['time'].corr(data['turbidity_YSI_80ft(NTU+)'], method='spearman'))
print('Correlación Kendall: ', data['time'].corr(data['turbidity_YSI_80ft(NTU+)'], method='kendall'))
print("-------------------------------------------")

# Tiempo VS Chloride 1100 ft (mg/L)
print("Tiempo VS Chloride 1100 ft (mg/L)")
print('Correlación Pearson: ', data['time'].corr(data['turbidity_YSI_1100ft(NTU+)'], method='pearson'))
print('Correlación Spearman: ', data['time'].corr(data['turbidity_YSI_1100ft(NTU+)'], method='spearman'))
print('Correlación Kendall: ', data['time'].corr(data['turbidity_YSI_1100ft(NTU+)'], method='kendall'))
print("-------------------------------------------")

# Tiempo VS Difference
print("Tiempo VS Difference")
print('Correlación Pearson: ', data['time'].corr(data['difference_turbidity'], method='pearson'))
print('Correlación Spearman: ', data['time'].corr(data['difference_turbidity'], method='spearman'))
print('Correlación Kendall: ', data['time'].corr(data['difference_turbidity'], method='kendall'))
```
