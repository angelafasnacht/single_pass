```python
# Se importan las librerias
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

### Se cargan los datos

```python
data = pd.read_csv("data/data_all.csv", sep=";")

data
```

### Preprocesamiento

```python
# Se buscan registros NaN para eliminarlos
print('Columna      Cantidad NaN')
print(data.isnull().sum(axis = 0))
print(data.shape)
```

```python
# Se genera las estadísticas del DataFrame
data.describe()
```

### -------------------------------------------------------------------------------

## Correlaciones

### -------------------------------------------------------------------------------

### Para los dos días

```python
# Gráfico de Dispersión
fig, ax = plt.subplots(1, 1, figsize=(10,8))
# Azul
ax.scatter(x= data['time'], y= data['ORP_YSI_80ft(mV)'], alpha= 0.8)
# Naranja
ax.scatter(x= data['time'], y= data['ORP_YSI_1100ft(mV)'], alpha= 0.8)

ax.set_xlabel('Time')
ax.set_ylabel('ORP (mV)')
```

```python
# Cálculo de correlaciones
print("-------------------------------------------")
print("     Para los dos días     ")
print("-------------------------------------------")
# Tiempo VS ORP 80ft (mV)
print("Tiempo VS ORP 80ft (mV)")
print('Correlación Pearson: ', data['time'].corr(data['ORP_YSI_80ft(mV)'], method='pearson'))
print('Correlación Spearman: ', data['time'].corr(data['ORP_YSI_80ft(mV)'], method='spearman'))
print('Correlación Kendall: ', data['time'].corr(data['ORP_YSI_80ft(mV)'], method='kendall'))
print("-------------------------------------------")

# Tiempo VS ORP 1100 ft (mV) 
print("Tiempo VS ORP 1100 ft (mV)")
print('Correlación Pearson: ', data['time'].corr(data['ORP_YSI_1100ft(mV)'], method='pearson'))
print('Correlación Spearman: ', data['time'].corr(data['ORP_YSI_1100ft(mV)'], method='spearman'))
print('Correlación Kendall: ', data['time'].corr(data['ORP_YSI_1100ft(mV)'], method='kendall'))
print("-------------------------------------------")

# Tiempo VS Difference 
print("Tiempo VS Difference")
print('Correlación Pearson: ', data['time'].corr(data['difference_ORP_YSI'], method='pearson'))
print('Correlación Spearman: ', data['time'].corr(data['difference_ORP_YSI'], method='spearman'))
print('Correlación Kendall: ', data['time'].corr(data['difference_ORP_YSI'], method='kendall'))
```
