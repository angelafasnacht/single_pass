```python
# Se importan las librerias
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

### Se cargan los datos

```python
data = pd.read_csv("data/data_all.csv", sep=";")

data
```

### Preprocesamiento

```python
# Se buscan registros NaN para eliminarlos
print('Columna      Cantidad NaN')
print(data.isnull().sum(axis = 0))
print(data.shape)
```

```python
# Se eliminan los registros NaN del DataFrame
data = data.dropna()
```

```python
# Se genera las estadísticas del DataFrame
data.describe()
```

## Correlaciones

### -------------------------------------------------------------------------------

### Para los dos días

```python
# Gráfico de Dispersión
fig, ax = plt.subplots(1, 1, figsize=(10,8))
# Azul
ax.scatter(x= data['time'], y= data['sp_cond_YSI_80ft(µS/cm)'], alpha= 0.8)
# Naranja
ax.scatter(x= data['time'], y= data['sp_cond_YSI_1100ft(µS/cm)'], alpha= 0.8)
ax.set_xlabel('Time')
ax.set_ylabel('Sp Cond (µS/cm)')
```

```python
# Cálculo de correlaciones
print("-------------------------------------------")
print("     Para las dos días     ")
print("-------------------------------------------")
# tiempo VS Sp Cond 80ft (µS/cm)
print("tiempo VS Sp Cond 80ft (µS/cm)")
print('Correlación Pearson: ', data['time'].corr(data['sp_cond_YSI_80ft(µS/cm)'], method='pearson'))
print('Correlación Spearman: ', data['time'].corr(data['sp_cond_YSI_80ft(µS/cm)'], method='spearman'))
print('Correlación Kendall: ', data['time'].corr(data['sp_cond_YSI_80ft(µS/cm)'], method='kendall'))
print("-------------------------------------------")

# time VS Sp Cond 1100 ft (µS/cm) 
print("time VS Sp Cond 1100 ft (µS/cm)")
print('Correlación Pearson: ', data['time'].corr(data['sp_cond_YSI_1100ft(µS/cm)'], method='pearson'))
print('Correlación Spearman: ', data['time'].corr(data['sp_cond_YSI_1100ft(µS/cm)'], method='spearman'))
print('Correlación Kendall: ', data['time'].corr(data['sp_cond_YSI_1100ft(µS/cm)'], method='kendall'))
print("-------------------------------------------")

# time VS Difference 
print("time VS Difference ")
print('Correlación Pearson: ', data['time'].corr(data['difference_sp_cond_YSI'], method='pearson'))
print('Correlación Spearman: ', data['time'].corr(data['difference_sp_cond_YSI'], method='spearman'))
print('Correlación Kendall: ', data['time'].corr(data['difference_sp_cond_YSI'], method='kendall'))
```
