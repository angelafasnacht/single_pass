## Correlaciones

### -------------------------------------------------------------------------------

### Para los dos días

```python
# Gráfico de Dispersión
fig, ax = plt.subplots(1, 1, figsize=(10,8))
# Azul
ax.scatter(x= data['time'], y= data['chloride_80ft(mg/L)'], alpha= 0.8)
# Naranja
ax.scatter(x= data['time'], y= data['chloride_1100ft(mg/L)'], alpha= 0.8)
ax.set_xlabel('Time')
ax.set_ylabel('Chloride(mg/L)')

#plt.savefig("chloride.png")
```

```python
# Cálculo de correlaciones
print("-------------------------------------------")
print("     Para los dos días     ")
print("-------------------------------------------")
# tiempo VS Chloride 80ft (mg/L)
print("tiempo VS Chloride 80ft (mg/L)")
print('Correlación Pearson: ', data['time'].corr(data['chloride_80ft(mg/L)'], method='pearson'))
print('Correlación Spearman: ', data['time'].corr(data['chloride_80ft(mg/L)'], method='spearman'))
print('Correlación Kendall: ', data['time'].corr(data['chloride_80ft(mg/L)'], method='kendall'))
print("-------------------------------------------")

# tiempo VS Chloride 1100 ft (mg/L) 
print("tiempo VS Chloride 1100 ft (mg/L)")
print('Correlación Pearson: ', data['time'].corr(data['chloride_1100ft(mg/L)'], method='pearson'))
print('Correlación Spearman: ', data['time'].corr(data['chloride_1100ft(mg/L)'], method='spearman'))
print('Correlación Kendall: ', data['time'].corr(data['chloride_1100ft(mg/L)'], method='kendall'))
print("-------------------------------------------")

# Tiempo VS Difference
print("Tiempo VS Difference")
print('Correlación Pearson: ', data['time'].corr(data['difference_chloride'], method='pearson'))
print('Correlación Spearman: ', data['time'].corr(data['difference_chloride'], method='spearman'))
print('Correlación Kendall: ', data['time'].corr(data['difference_chloride'], method='kendall'))
```
