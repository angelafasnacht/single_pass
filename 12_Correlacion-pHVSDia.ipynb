### Se importan las librerias

```python
# Se importan las librerias
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

### Se cargan los datos

```python
data = pd.read_csv("data/data_all.csv", sep=";")

data
```

### Preprocesamiento

```python
# Se buscan registros NaN para eliminarlos
print('Columna      Cantidad NaN')
print(data.isnull().sum(axis = 0))
print(data.shape)
```

```python
# Se eliminan los registros NaN del DataFrame
data = data.dropna()
```

```python
# Se genera las estadísticas del DataFrame
data.describe()
```

```python
# Gráfica unificada de la variable
variable1 = data['pH_YSI_80ft']
variable2 = data['pH_YSI_1100ft']

# pH 80ft --> Azul
plot = variable1.plot(figsize=(10, 8))
# pH 1100ft --> Naranja
plot = variable2.plot(figsize=(10, 8))
```

```python
# Se ordena el DataFrame de mayor a menor valor 1100ft (para los dos días)
data_1100 = data.drop(['pH_YSI_80ft'], axis=1)

pH_1100 = data_1100.sort_values('pH_YSI_1100ft')

print(pH_1100)
```

### -------------------------------------------------------------------------------

## Correlaciones

### -------------------------------------------------------------------------------

### Para los dos días

```python
# Gráfico de Dispersión
fig, ax = plt.subplots(1, 1, figsize=(10,8))
# Azul
ax.scatter(x= data['time'], y= data['pH_YSI_80ft'], alpha= 0.8)
# Naranja
ax.scatter(x= data['time'], y= data['pH_YSI_1100ft'], alpha= 0.8)
ax.set_xlabel('Time')
ax.set_ylabel('pH (ft)')
```

```python
# Cálculo de correlaciones
print("-------------------------------------------")
print("     Para los dos días     ")
print("-------------------------------------------")
# tiempo VS pH 80ft VS
print("tiempo VS pH 80ft VS")
print('Correlación Pearson: ', data['time'].corr(data['pH_YSI_80ft'], method='pearson'))
print('Correlación Spearman: ', data['time'].corr(data['pH_YSI_80ft'], method='spearman'))
print('Correlación Kendall: ', data['time'].corr(data['pH_YSI_80ft'], method='kendall'))
print("-------------------------------------------")

# tiempo VS pH 1100ft VS 
print("tiempo VS pH 1100ft VS")
print('Correlación Pearson: ', data['time'].corr(data['pH_YSI_1100ft'], method='pearson'))
print('Correlación Spearman: ', data['time'].corr(data['pH_YSI_1100ft'], method='spearman'))
print('Correlación Kendall: ', data['time'].corr(data['pH_YSI_1100ft'], method='kendall'))
print("-------------------------------------------")

# tiempo VS Difference 
print("tiempo VS Difference ")
print('Correlación Pearson: ', data['time'].corr(data['difference_pH'], method='pearson'))
print('Correlación Spearman: ', data['time'].corr(data['difference_pH'], method='spearman'))
print('Correlación Kendall: ', data['time'].corr(data['difference_pH'], method='kendall'))
```
