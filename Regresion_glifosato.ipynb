## Modelo de Regresión Glifosato

### Se importan las librerias

```python
# Se importan las librerias
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
#from mpl_toolkits.mplot3d import Axes3D
#from matplotlib import cm

# Modelos de Regresión: árbol de regresión, SVR y KNN
# Se importa la libreria de árbol de regresión
from sklearn.tree import DecisionTreeRegressor 
# Se importa la libreria de SVR
from sklearn.svm import SVR
# Se importa la libreria de KNN de regresión
from sklearn.neighbors import KNeighborsRegressor

# Modelos de Regresión (ensamble): Random Forest, Gradient Boosting, Ada Boost
# Se importa la libreria de Random Forest
from sklearn.ensemble import RandomForestRegressor
# Se importa la libreria de Gradient Boosting
from sklearn.ensemble import GradientBoostingRegressor
# Se importa la libreria de Ada Boost
from sklearn.ensemble import AdaBoostRegressor

# Se importa la libreria para dividir los datos de entrenamiento y de pruebas
from sklearn.model_selection import train_test_split
# Se importa la libreria para la métrica R cuadrado y RMSE 
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

### Se cargan los datos

```python
data = pd.read_csv("data/data_all_glyphosate.csv", sep=";")

data
```

### Preprocesamiento

```python
# Se buscan registros NaN para eliminarlos
print('Columna                Cantidad NaN')
print(data.isnull().sum(axis = 0))
print(data.shape)
```

```python
# Se eliminan los registros NaN del DataFrame
#data = data.dropna()
```

```python
# Se genera las estadísticas del DataFrame
data.describe()
```

### Predicción: Todas las características

```python
# Se definen los valores de las características:
features = ['chloride_80ft(mg/L)','chloride_1100ft(mg/L)',
                   'CL2_80ft(PPM)', 'CL2_1100ft(PPM)',
                   'DO_YSI_80ft(mg/L)','DO_YSI_1100ft(mg/L)',
                   'FreeClorine_ATI_80ft(PPM)','FreeClorine_ATI_1100ft(PPM)',
                   'FreeClorine_X_80ft(PPM)','FreeClorine_X_1100ft(PPM)',
                   'ORP_YSI_80ft(mV)','ORP_YSI_1100ft(mV)',
                   'pH_YSI_80ft','pH_YSI_1100ft',
                   'sp_cond_YSI_80ft(µS/cm)','sp_cond_YSI_1100ft(µS/cm)',
                   'turbidity_YSI_80ft(NTU+)','turbidity_YSI_1100ft(NTU+)']
                   
X = data[features]
# Variable Objetivo: grade
y = data['concentration']

# Se dividen los datos para el entrenamiento (80% entrenamiento y 20% pruebas)
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, random_state = 1)

# Se crea el modelo
arbol_regresion = DecisionTreeRegressor() 
  
# Se entrena el modelo
arbol_regresion.fit(X_train, y_train)

predicciones = arbol_regresion.predict(X = X_test)

# Se utiliza la métrica: R2
print('R2-árbol de regresión:',  r2_score(y_test, predicciones))

# Se utiliza la métrica: raíz del error cuadrático medio (rmse)
rmse = mean_squared_error( y_true  = y_test, y_pred  = predicciones, squared = False)

print('RMSE-árbol de regresión:', rmse)

# Se utiliza la métrica: MAE
mae = mean_absolute_error(y_test, predicciones)

print('MAE-árbol de regresión:', mae)
```

### Se extraen las mejores características (Best Features)

```python
# Se extraen las mejores características para le modelo
best_features = pd.DataFrame({'feature':X_train.columns,'Importancia':np.round(arbol_regresion.feature_importances_,3)})
best_features = best_features.sort_values('Importancia',ascending=False).set_index('feature')
best_features.head()

```

##### DecisionTreeRegressor

```python
# DecisionTreeRegressor: Ajuste Hiperparámetros (Random Search)

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import RepeatedKFold
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import RandomizedSearchCV

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

dtr_ah = DecisionTreeRegressor()

# 
criterion = ["squared_error", "friedman_mse", "absolute_error", "poisson"]
# 
splitter = ["best", "random"]
#
max_depth = [2, 5, 10, 15, 20]
#
min_samples_split = [2, 5, 10, 15, 20]
#
min_samples_leaf = [1, 2, 4, 6]


grid = dict(criterion = criterion,
            splitter = splitter,
            max_depth = max_depth,
            min_samples_split = min_samples_split,
            min_samples_leaf = min_samples_leaf
           )

#print("[INFO] grid searching over the hyperparameters...")
cvFold = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)
randomSearch = RandomizedSearchCV(estimator= dtr_ah,
                                  n_jobs=-1,
                                  cv=cvFold,
                                  param_distributions=grid,
                                  scoring="neg_mean_squared_error")
searchResults = randomSearch.fit(X_train, y_train.ravel())

# extract the best model and evaluate it
#print("[INFO] evaluating...")
bestModel = searchResults.best_estimator_

print("Best Parameters (RandomSearch):", bestModel)
print("-----------------------------------------------------------")

# Se crea un objeto con los mejores ajustes de Hiperparámetros
DTR_AH = bestModel
  
# Se entrena el modelo con los mejores parámetros
DTR_AH.fit(X_train, y_train)

predictDTR_AH = DTR_AH.predict(X = X_test)

#R2
r2DTR_AH= r2_score(y_test, predictDTR_AH)

#RMSE
rmseDTR_AH = mean_squared_error(y_true = y_test, y_pred = predictDTR_AH, squared=False)

#MAE
maeDTR_AH = mean_absolute_error(y_test, predictDTR_AH)

print('R2 = %.2f' % r2DTR_AH)
print('RMSE = %.2f' % rmseDTR_AH)
print('MAE = %.2f' % maeDTR_AH)
```

##### SVR

```python
# SVR: Ajuste Hiperparámetros (Random Search)

from scipy.stats import loguniform

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

svr = SVR()
kernel = ["linear", "rbf", "sigmoid", "poly"]
tolerance = loguniform(1e-6, 1e-3)
C = [1, 1.5, 2, 2.5, 3]
grid = dict(kernel=kernel, tol=tolerance, C=C)

cvFold = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)
randomSearch = RandomizedSearchCV(estimator= svr, n_jobs=-1,
	cv=cvFold, param_distributions=grid,
	scoring="neg_mean_squared_error")
searchResults = randomSearch.fit(X_train, y_train)

# extract the best model and evaluate it
bestModel = searchResults.best_estimator_

print("Best Parameters (RandomSearch):", bestModel)
print("-----------------------------------------------------------")

# Se crea un objeto con los mejores ajustes de Hiperparámetros
SVR_AH = bestModel

# Se entrena el modelo con los mejores parámetros
SVR_AH.fit(X_train, y_train)

predictSVR_AH = SVR_AH.predict(X_test)

#R2
r2SVR_AH= r2_score(y_test, predictSVR_AH)

#RMSE
rmseSVR_AH = mean_squared_error(y_true = y_test, y_pred = predictSVR_AH, squared=False)

#MAE
maeSVR_AH = mean_absolute_error(y_test, predictSVR_AH)

print('R2 = %.2f' % r2SVR_AH)
print('RMSE = %.2f' % rmseSVR_AH)
print('MAE = %.2f' % maeSVR_AH)
```

#### KNN Regressor

```python
# KNN-Regressor: Ajuste Hiperparámetros (Random Search)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

knn = KNeighborsRegressor()

# 
n_neighbors = [1, 2, 5, 10]
# 
weights = ["uniform", "distance"]
#
algorithm = ["auto", "ball_tree", "kd_tree", "brute"]
#
leaf_size = [5, 10, 20, 30, 40, 50]
#
p = [2]

grid = dict(n_neighbors = n_neighbors,
            weights = weights,
            algorithm = algorithm,
            leaf_size = leaf_size,
            p = p
           )

#print("[INFO] grid searching over the hyperparameters...")
cvFold = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)
randomSearch = RandomizedSearchCV(estimator=knn,
                                  n_jobs=-1,
                                  cv=cvFold,
                                  param_distributions=grid,
                                  scoring="neg_mean_squared_error")
searchResults = randomSearch.fit(X_train, y_train.ravel())

# extract the best model and evaluate it
bestModel = searchResults.best_estimator_

print("Best Parameters (RandomSearch):", bestModel)
print("-----------------------------------------------------------")

knn = bestModel

knn = knn.fit(X_train, y_train)

predictKNN = knn.predict(X = X_test)

#R2
r2KNN= r2_score(y_test, predictKNN)

#RMSE
rmseKNN = mean_squared_error(y_true = y_test, y_pred = predictKNN, squared=False)

#MAE
maeKNN = mean_absolute_error(y_test, predictKNN)

print('R2 = %.2f' % r2KNN)
print('RMSE = %.2f' % rmseKNN)
print('MAE = %.2f' % maeKNN)
```

#### --------------- Algoritmos de Emsamble ---------------

##### RamdomForestRegressor

```python
# RamdomForestRegressor: Ajuste Hiperparámetros (Random Search)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

rfr = RandomForestRegressor()

# Número de árboles en un bosque aleatorio
n_estimators = [int(x) for x in np.linspace(start=20, stop=200, num = 10)]
# Número de características a considerar en cada división
max_features = [1.0]
# Número máximo de niveles en el árbol
max_depth = [int(x) for x in np.linspace(10, 110, num = 11)]
max_depth.append(None)
# Número mínimo de muestras requeridas para dividir un nodo
min_samples_split = [2, 5, 10, 15, 20]
# Número mínimo de muestras requeridas en cada nodo hoja
min_samples_leaf = [1, 2, 4, 6]
# Método de selección de muestras para entrenar cada árbol 
bootstrap = [True, False]
grid = dict(n_estimators=n_estimators,
            max_features=max_features,
            max_depth=max_depth,
            min_samples_split=min_samples_split,
            min_samples_leaf=min_samples_leaf,
            bootstrap=bootstrap)

cvFold = RepeatedKFold(n_splits=5, n_repeats=3, random_state=1)
randomSearch = RandomizedSearchCV(estimator= rfr,
                                  n_jobs=-1,
                                  cv=cvFold,
                                  param_distributions=grid,
                                  scoring="neg_mean_squared_error")
searchResults = randomSearch.fit(X_train, y_train.ravel())


# extract the best model and evaluate it
bestModel = searchResults.best_estimator_

print("Best Parameters (RandomSearch):", bestModel)
print("-----------------------------------------------------------")

# Se crea un objeto con los mejores ajustes de Hiperparámetros
RFR_AH = bestModel

# Se entrena el modelo
RFR_AH.fit(X_train, y_train)

# Se entrena el modelo con los mejores parámetros
predictRFR_AH = RFR_AH.predict(X = X_test)

#R2
r2RFR_AH= r2_score(y_test, predictRFR_AH)

#RMSE
rmseRFR_AH = mean_squared_error(y_true = y_test, y_pred = predictRFR_AH, squared=False)

#MAE
maeRFR_AH = mean_absolute_error(y_test, predictRFR_AH)

print('R2 = %.2f' % r2RFR_AH)
print('RMSE = %.2f' % rmseRFR_AH)
print('MAE = %.2f' % maeRFR_AH)
```

#### Gradient Boosting Regressor

```python
# GradientBoostingRegressor: Ajuste Hiperparámetros (Random Search)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

gb = GradientBoostingRegressor()

# 
loss = ["squared_error", "absolute_error", "huber", "quantile"]
# 
learning_rate = [0.1]
#
n_estimators = [10, 50, 100, 200]
#
subsample = [1.0]
#
criterion = ["friedman_mse", "squared_error"]
#
min_samples_split = [2]
#
min_samples_leaf = [1]
#
max_depth = [3]

grid = dict(loss = loss,
            learning_rate = learning_rate,
            n_estimators = n_estimators,
            subsample = subsample,
            criterion = criterion,
            min_samples_split = min_samples_split,
            min_samples_leaf = min_samples_leaf,
            max_depth = max_depth)

#print("[INFO] grid searching over the hyperparameters...")
cvFold = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)
randomSearch = RandomizedSearchCV(estimator=gb,
                                  n_jobs=-1,
                                  cv=cvFold,
                                  param_distributions=grid,
                                  scoring="neg_mean_squared_error")
searchResults = randomSearch.fit(X_train, y_train.ravel())

# extract the best model and evaluate it
bestModel = searchResults.best_estimator_

print("Best Parameters (RandomSearch):", bestModel)
print("-----------------------------------------------------------")

GB = bestModel.fit(X_train, y_train)

predictGB = GB.predict(X = X_test)

#R2
r2GB= r2_score(y_test, predictGB)

#RMSE
rmseGB = mean_squared_error(y_true = y_test, y_pred = predictGB, squared=False)

#MAE
maeGB = mean_absolute_error(y_test, predictGB)

print('R2 = %.2f' % r2GB)
print('RMSE = %.2f' % rmseGB)
print('MAE = %.2f' % maeGB)
```

#### Ada Boost Regressor

```python
# AdaBoostRegressor: Ajuste Hiperparámetros (Random Search)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

ab = AdaBoostRegressor()

# 
n_estimators = [10, 20, 30, 40, 50]
# 
learning_rate = [1.0]
#
loss = ["linear", "square", "exponential"]

grid = dict(n_estimators = n_estimators,
            learning_rate = learning_rate,
            loss = loss)

#print("[INFO] grid searching over the hyperparameters...")
cvFold = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)
randomSearch = RandomizedSearchCV(estimator=ab,
                                  n_jobs=-1,
                                  cv=cvFold,
                                  param_distributions=grid,
                                  scoring="neg_mean_squared_error")
searchResults = randomSearch.fit(X_train, y_train.ravel())

# extract the best model and evaluate it
bestModel = searchResults.best_estimator_

print("Best Parameters (RandomSearch):", bestModel)
print("-----------------------------------------------------------")

AB = bestModel.fit(X_train, y_train)

predictAB = AB.predict(X = X_test)

#R2
r2AB= r2_score(y_test, predictAB)

#RMSE
rmseAB = mean_squared_error(y_true = y_test, y_pred = predictAB, squared=False)

#MAE
maeAB = mean_absolute_error(y_test, predictAB)

print('R2 = %.2f' % r2AB)
print('RMSE = %.2f' % rmseAB)
print('MAE = %.2f' % maeAB)
```
